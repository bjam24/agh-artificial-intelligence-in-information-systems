## Przygotowanie modelu 

```python
import shap

data, labels = shap.datasets.adult()
```

```python
# zapoznaj się z danymi
data
```

```python
# zapoznaj sie z danymi
labels.shape
```

```python
import lightgbm as lgbm
params = {'objective':'binary',
          'booster_type':'gbdt',
          'max_depth':6,
          'learning_rate':0.05,
          'metric':'auc'
    }

dtrain = lgbm.Dataset(data,labels)

model = lgbm.train(params,dtrain,num_boost_round=100,valid_sets=[dtrain])

model.save_model('saved_model.txt')
```
